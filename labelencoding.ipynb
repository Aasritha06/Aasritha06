```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
data={"color":["Red","Green","Yellow","Orange"]}
df=pd.DataFrame(data)
df
```

```python
label_encoder=LabelEncoder()
df['Color_label']=label_encoder.fit_transform(df['color'])
df.head()
```

```python
one_hot_encoder=OneHotEncoder()
one_hot_encoded=one_hot_encoder.fit_transform(df[['color']])
one_hot_encoded_data=pd.DataFrame(one_hot_encoded.toarray(),columns=one_hot_encoder.get_feature_names_out(['color']))
df=pd.concat([df,one_hot_encoded_data],axis=1)
df
```
